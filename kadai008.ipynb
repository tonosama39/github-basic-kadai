```python
from sklearn.datasets import load_wine
```

```python
dataset = load_wine()
dataset
```

```python
dataset.data
```

```python
dataset.feature_names
```

```python
import pandas as pd
df = pd.DataFrame(data = dataset.data, columns = dataset.feature_names)
df.head()
```

```python
dataset.target
```

```python
df['category'] = dataset.target
df.tail()
```

```python
winef = dataset.data
winetype = dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(winef,winetype, test_size = 0.3, random_state = 5)
```

```python
winef_train, winef_test, winetype_train, winetype_test = train_test_split(winef,winetype, test_size = 0.3, random_state = 5)
```

```python
print(winef_train.shape)
print(winef_test.shape)
print(winetype_train.shape)
print(winetype_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state = 3)
```

```python
model.fit(winef_train, winetype_train)
```

```python
winetype_pred = model.predict(winef_test)
winetype_pred
```

```python
winetype_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(winetype_test, winetype_pred)
```

```python
model.score(winef_test, winetype_test)
```
